```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://www.billboard.com/charts/hot-100/"
```

```python
def scrap_hot100(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    print("Status code: " + str(response.status_code))
    title = []
    artists = []
    for i in soup.select("li.o-chart-results-list__item > h3"):
        title.append(i.get_text().strip())
    for i in soup.select("li.o-chart-results-list__item > span:nth-child(2)"):
        artists.append(i.get_text().strip())
    unwanted = {'NEW', 'RE-\nENTRY'}
    artists = [ele for ele in artists if ele not in unwanted]
    songs = {'title': title,
            'artist': artists}
    hot100 = pd.DataFrame(songs).reset_index(drop = True)
    hot100.to_csv('hot100_songs.csv', index=False)
    return hot100
```

```python
scrap_hot100('https://www.billboard.com/charts/hot-100')
```

```python
def check_duplicates(df_nothot, df_hot):
    title = []
    artist = []
    nots = list(zip(df_nothot['title'], df_nothot['artist']))
    hots = list(zip(df_hot['title'], df_hot['artist']))
    for elem in nots:
        if elem not in hots:
            title.append(elem[0])
            artist.append(elem[1])
    #df = pd.DataFrame()
    df_nothot['title'] = title
    df_nothot['artist'] = artist
    
    return df_nothot.reset_index(drop = True)
```
